```python
#PERSIAPAN
import pandas as pd # untuk membaca data
import numpy as np # untuk mengolah data
import matplotlib.pyplot as plt # membuat grafik
import seaborn as sns # untuk menentukan hubungan antara dua variabel
```

```python
# PREPROCESS
# membaca data csv yang di importkan
mush = pd.read_csv('./dataset/mushrooms.csv')
mush.head()
```

```python
# Mengubah data dari kolom menjadi baris, guna melihat keseluruhan data
mush.head().transpose()
```

```python
# PROSES TRAINING
# Mengimport LabelEncoder untuk mengkonversi nilai kategori menjadi urutan
from sklearn.preprocessing import LabelEncoder

# Memanggil model label encoder dalam variabel le
le = LabelEncoder()

# Melakukan perulangan pada setiap kolom agar terkonversi semuanya
# Menggunakan fungsi fit_transform untuk menyesuaikan data dan mengubah baris menjadi kolom maupun sebaliknya 
for column in mush.columns:
    mush[column] = le.fit_transform(mush[column])

# Memilih kolom data untuk sumbu x 
x = mush.drop(["class"], axis = 1)

# Memilih kolom data untuk sumbu y
y = mush['class']
```

```python
# mengimport model train_test_split untuk ngesplit sebagian data train menjadi data testing
from sklearn.model_selection import train_test_split

# Data testing diambil dari 20% data training menggunakan 'test_size=0.2'
# Data testing diambil dari 1 kolom dari data training menggunakan 'random_state=1'
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)
```

```python
# Mengimpor klasifikasi Decision tree
from sklearn.tree import DecisionTreeClassifier

# Memanggil model decision tree (dt) dan memiliki parameter kedalaman max = 6 dan random state/kolom = 1
dt = DecisionTreeClassifier(max_depth=6, random_state=1)

# Menggunakan fungsi fit untuk menyesuaikan model dt dengan data training
dt.fit(X_train, y_train)

# Menentukan hasil prediksi dari x_test
y_pred = dt.predict(X_test)

# Print data test hasil prediksi data test dengan 1 baris dan 5 kolom
print(y_pred[0:5])
```

```python
# MENGEVALUASI CLASSIFICATION TREE

# Mengimport matriks accuracy score
from sklearn.metrics import accuracy_score

# Memprediksi test set labels
y_pred = dt.predict(X_test)

# Menghitung set accuracy
acc = accuracy_score(y_test, y_pred)
print("Test set accuracy: {:.2f}".format(acc))
```

```python
# MENGGUNAAN ENTROPY SEBAGAI KRITERIA
# menggunakan keseluruhan dari 30 fitur dalam dataset, yang dibagi menjadi 80% untuk train dan 20% untuk test.

# Mengimport tree untuk klasifikasi decision tree
from sklearn.tree import DecisionTreeClassifier

# Memanggil model decision tree (dt_entropy) dan memiliki parameter kedalaman max = 8, menggunakan kriteria entropy dan random state/kolom = 1
dt_entropy = DecisionTreeClassifier(max_depth=8, criterion='entropy', random_state=1)

# fit atau sesuaikan dt_entropy kedalam training set
dt_entropy.fit(X_train, y_train)
```

```python
# MENGGUNAKAN GIN SEBAGAI KRITERIA

# Memanggil model decision tree (dt_entropy) dan memiliki parameter kedalaman max = 8, menggunakan kriteria gini dan random state/kolom = 1
dt_gini = DecisionTreeClassifier(max_depth=8, criterion='gini', random_state=1)

# fit atau sesuaikan dt_gini kedalam training set
dt_gini.fit(X_train, y_train)
```

```python
# Mengimpor matriks untuk hasil akurasi
from sklearn.metrics import accuracy_score

# Menggunakan dt_entropy dan gin untuk memprediksi test set labels
y_pred = dt_entropy.predict(X_test)
y_pred_gini = dt_gini.predict(X_test)

# Mengevaluasi accuracy_scor edari entropy dan gin
accuracy_entropy = accuracy_score(y_test, y_pred)
accuracy_gini = accuracy_score(y_test, y_pred_gini)

# Print accuracy_entropy
print("Accuracy achieved by using entropy: ", accuracy_entropy)

# Print accuracy_gini
print("Accuracy achieved by using gini: ", accuracy_gini)
```
