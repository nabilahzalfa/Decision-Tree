```python
# DECISION TREE UNTUK REGRESSION
import pandas as pd # untuk membaca data
import numpy as np # untuk mengolah data
import matplotlib.pyplot as plt # membuat grafik
```

```python
# PREPROCESS
# membaca data csv yang di importkan
mpg = pd.read_csv('./dataset/auto.csv')
mpg.head()
```

```python
# Membaca data dengan menggunakan dummies
# Dummies digunakan untukmengkonvert/mengubah variabel kategori ke dalam variabel numerik
# Dengan melakukan proses One-Hot-Encode terhadap variabel kategori
mpg = pd.get_dummies(mpg)
mpg.head()
```

```python
# Mengubah data dari kolom menjadi baris
mpg.head().transpose()
```

```python
# Mengimport model train_test_split untuk ngesplit sebagian data train menjadi data testing
from sklearn.model_selection import train_test_split

# Memilih kolom data suntuk sumbu x dengan menghapus kolom mpg
X = mpg.drop('mpg', axis='columns')

# Memilih kolom data untuk sumbu y berisi seluruh data mpg
y = mpg['mpg']

# Data testing diambil dari 20% data training menggunakan 'test_size=0.2'
# Data testing diambil dari 1 kolom dari data training menggunakan 'random_state=3'
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=3)
```

```python
from sklearn.tree import DecisionTreeRegressor

# Memanggil model Decision Tree Regression (dt)
# Dengan kedalaman = 8, jumlah minimum sampel yang diperlukan berada pada leaf node = 0.13 atau 13%, dan kolom yang diambil = 3
dt = DecisionTreeRegressor(max_depth=8, min_samples_leaf=0.13, random_state=3)

# Menggunakan fungsi fit untuk menyesuaikan model dt dengan data training
dt.fit(X_train, y_train)
```

```python
# MENGEVALUASI REGRESSION TREE
# RMSE model mengukur rata-rata, seberapa besar prediksi model yang dihasilkan berbeda dari label yang sebenarnya.
# RMSE model dapat diperoleh dengan menghitung akar kuadrat dari model Mean Squared Error (MSE).

# Mengimport matriks model mean squared error
from sklearn.metrics import mean_squared_error

# Menentukan hasil prediksi dari x_test
y_pred = dt.predict(X_test)

# Menghitung mse
mse_dt = mean_squared_error(y_test, y_pred)

# Menghitung root MSE
# RSME diperoleh dari akar kuadrat dari MSE
rmse_dt = mse_dt ** (1/2)

# Print hasil RMSE
print("Test set RMSE dt: {:.2f}".format(rmse_dt))
```

```python
# Mengimport model linear regression
from sklearn.linear_model import LinearRegression

# Memanggil fungsi linear dengan variabel lr
lr = LinearRegression()

# fit atau sesuaikan pada model lr
lr.fit(X_train, y_train)
```

```python
# Memprediksi linier regression
y_pred_lr = lr.predict(X_test)

# Menghitng MSE model lr
mse_lr = mean_squared_error(y_test, y_pred_lr)

# Menghitung RMSE model lr diperoleh dari akar kuadrat dari MSE model Linear Regression
rmse_lr = mse_lr ** 0.5

# Print rmse_lr
print("Linear Regression test set RMSE: {:.2f}".format(rmse_lr))

# Print rmse_dt
print("Regression Tree test set RMSE: {:.2f}".format(rmse_dt))
```
